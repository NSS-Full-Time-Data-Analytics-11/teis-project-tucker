```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
colors = ["#C37681", "#EAE086", "#838385", "#89C1DF", "#5BAC82"]
sns.set_palette(sns.color_palette(colors))
```

```python
teis  = pd.read_csv('./data/teis.csv', dtype = {"Child ID" : object, "Cognitive T-Score" : object})
teis
```

```python
teis['Program Label'].value_counts()
```

```python
pulls = teis[['Child ID',
 'Gender',
 'Date of Birth',
 'Location - Sub Level 1',
 'Program Label','Adaptive Developmental Quotient',
 'Adaptive Percentile Rank','Social-Emotional Developmental Quotient',
 'Social-Emotional Percentile Rank','Communication Developmental Quotient',
 'Communication Percentile Rank','Motor Developmental Quotient',
 'Motor Percentile Rank','Cognitive Developmental Quotient',
 'Cognitive Percentile Rank','BDI-3 Total Sum of Scaled Scores',
 'BDI-3 Total Developmental Quotient',
 'BDI-3 Total Percentile Rank',
 'Adaptive-Self Care Examiner',
 'Adaptive-Self Care Date of Testing']]
pulls
```

```python
pulls['Location - Sub Level 1'].value_counts()
```

```python
def Region(x):
    if x in ('East Tennessee', 'First Tennessee', 'Southeast Tennessee'):
        return 'East TN'
    elif x in ('Greater Nashville', 'Upper Cumberland', 'South Central'):
        return 'Middle TN'
    elif x in ('Northwest', 'Southwest', 'Memphis Delta'):
        return 'West TN'
pulls['Region'] = pulls['Location - Sub Level 1'].apply(Region)
pulls.head()
```

```python
pulls.groupby('Region')['Adaptive Developmental Quotient'].describe()
```

```python
pulls.plot(x = 'Region', y = 'Adaptive Developmental Quotient', kind = 'box')
```

```python
pulls.boxplot(column ='Adaptive Percentile Rank', by='Region')
pulls.boxplot(column ='Social-Emotional Percentile Rank', by='Region')
pulls.boxplot(column ='Communication Percentile Rank', by='Region')
pulls.boxplot(column ='Motor Percentile Rank', by='Region')
pulls.boxplot(column ='Cognitive Percentile Rank', by='Region')
pulls.boxplot(column ='BDI-3 Total Percentile Rank', by='Region')
```

```python
sns.boxplot(x = pulls['Region'], y = pulls['Adaptive Developmental Quotient'])
```

```python
adaptive = ['#C37681']
cognitive = ['#EAE086']
communication = ['#838385']
motor = ['#89C1DF']
socialemo = ['#5BAC82']
bdi3total = ['#F1B87E']
```

```python
fig, axes = plt.subplots(2, 3, figsize = (30, 12))


sns.boxplot(ax = axes[0,0], x = pulls['Location - Sub Level 1'], y = pulls['Adaptive Percentile Rank'], palette = adaptive).set(xticklabels=[], xlabel=None) 
sns.boxplot(ax = axes[0,1], x = pulls['Location - Sub Level 1'], y = pulls['Cognitive Percentile Rank'], palette = cognitive).set(xticklabels=[], xlabel=None) 
sns.boxplot(ax = axes[0,2], x = pulls['Location - Sub Level 1'], y = pulls['Communication Percentile Rank'], palette = communication).set(xticklabels=[], xlabel=None) 
motorplot = (sns.boxplot(ax = axes[1,0], x = pulls['Location - Sub Level 1'], y = pulls['Motor Percentile Rank'], palette = motor))
seplot = (sns.boxplot(ax = axes[1,1], x = pulls['Location - Sub Level 1'], y = pulls['Social-Emotional Percentile Rank'], palette = socialemo))
sns.boxplot(ax = axes[1,2], x = pulls['Location - Sub Level 1'], y = pulls['BDI-3 Total Percentile Rank'], palette = bdi3total)

plt.xticks(rotation = 90, fontsize = 16.5)
motorplot.set_xticklabels(motorplot.get_xticklabels(), rotation = 90, fontsize = 16.5)
seplot.set_xticklabels(seplot.get_xticklabels(), rotation = 90, fontsize = 16.5)
```
